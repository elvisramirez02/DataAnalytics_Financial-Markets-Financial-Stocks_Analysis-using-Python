### What is the present value of $418 received in 12 years with an annual interest rate of 2.2%?

```python
### Create a python function for calcule present value, and reuse the calculate as well

def presentValue(futureValue, interestRate, years):
    discountRate = 1 + (interestRate / 100)
    presentValue = futureValue / (discountRate ** years)
    return presentValue
```

```python
## Define the values to calcule present value using python function
futureValue = 418
interestRate = 2.2
years = 12

pv = presentValue(futureValue, interestRate, years)
print("The present value of: $",futureValue,"with an annual interest rate of",interestRate, "is $", round(pv, 2))
```

### If you deposit $2,400 in the bank for 15 years and the annual interest rate us 1.75%, what is the future value of your deposit in 15 years?

```python
### Create a python function for calcule future value, and reuse the calculate as well

def futureValue(presentValue, interestRate, years):
    futureValue = presentValue * (1 + interestRate)**years
    return futureValue
```

```python
## Define the values to calcule future value using python function

presentValue = 2400
interestRate = 0.0175
years = 15

fv = futureValue(presentValue, interestRate, years)
print("The future value of deposit: $",presentValue,"in 15 years, with an annual interest rate of:",round(interestRate*100, 2),"% will be: $", round(fv, 2))
```

###  Evaluate all 4 cases and calculate their NPV. Based on their NPV which case would you choose?
Case 1:

Case 2:

```python
### Create a python function for calcule Net present value, and reuse the calculate as well
def calculate_npv(cashflows, discount_rate):
    npv = 0
    for t in range(len(cashflows)):
        npv += cashflows[t] / (1 + discount_rate)**t
    return npv
```

```python
## Case 1: Initial Investment = $500 Cash inflows = 50,50,50,50,50,50,50,50 Discount rate = 12.5%
case1_initInvestment = 500
case1_cashFlows = [50, 50, 50, 50, 50, 50, 50, 50]
case1_discountRate = 0.125  # 12.5%

npv_case1 = calculate_npv([case1_initInvestment] + case1_cashFlows, case1_discountRate)

print("The Net Present Value (NPV) Case 1 is:", round(npv_case1, 2))
```

```python
## Case 2: Initial Investment = $350 Cash inflows = 57,61,25,5,18,35,99,100 Discount rate = 5.3%
case2_initInvestment = 350
case2_cashFlows = [57,61,25,5,18,35,99,100]
case2_discountRate = 0.053  # 5.3%

npv_case2 = calculate_npv([case2_initInvestment] + case2_cashFlows, case2_discountRate)

print("The Net Present Value (NPV) Case 2 is:", round(npv_case2, 2))
```

```python
## Case 3: Initial Investment = $180 Cash inflows = 20,20,20,20,20 Discount rate = 3.2%
case3_initInvestment = 180
case3_cashFlows = [20,20,20,20,20]
case3_discountRate = 0.032  # 3.2%

npv_case3 = calculate_npv([case3_initInvestment] + case3_cashFlows, case3_discountRate)

print("The Net Present Value (NPV) Case 3 is:", round(npv_case3, 2))
```

```python
## Case 4: Case 4: Initial Investment = $3800 Cash inflows = 720,320,650,480,550,180,1000,1200 Discount rate = 7.9%
case4_initInvestment = 3800
case4_cashFlows = [720,320,650,480,550,180,1000,1200]
case4_discountRate = 0.079  # 7.9%

npv_case4 = calculate_npv([case4_initInvestment] + case4_cashFlows, case4_discountRate)

print("The Net Present Value (NPV) Case 4 is:", round(npv_case4, 2))
```

relationship between decision trees and Random Forest? What are some of the pros/cons of each approach?


Decision trees and Random Forest are both machine learning algorithms used for classification and regression tasks. Random Forest is an ensemble method built on top of decision trees.

### 1. Decision Trees:

a) Decision trees are a basic machine learning technique that may be used for classification and regression applications.
b) They operate by recursively subdividing the data into subsets depending on the most informative characteristic at each stage, resulting in a tree-like structure that leads to a final conclusion or prediction.
- Because decision trees are simple to read and visualise, they may help you understand how the model makes decisions.
c) They can handle both numerical and categorical data without requiring much data preprocessing.
d) Decision trees are prone to overfitting, especially when the tree becomes deep and complex, which can lead to poor generalization on new data.

### Pros and Cons of Decision Trees:

#### Pros:
- Easy to interpret and visualize.
- Handles both numerical and categorical data.
- Requires minimal data preprocessing.
- Can capture non-linear relationships in the data.

####Cons:
- Prone to overfitting, especially with deep and complex trees.
- Sensitive to small variations in the data, leading to high variance.
- Lack of generalization on unseen data.


### 2. Random Forest:

Random Forest is an ensemble learning technique that builds multiple decision trees and combines their predictions to make a final prediction.
a) It works by creating multiple decision trees on different subsets of the data (randomly selected with replacement) and features (randomly selected without replacement).
b) The final prediction in Random Forest is typically the majority vote for classification problems or the average prediction for regression problems.
c) Random Forest tends to be more robust and accurate than a single decision tree because it reduces overfitting by combining predictions from multiple trees.
d) It can handle large datasets and high-dimensional feature spaces effectively.
The trade-off for improved performance is a loss of interpretability compared to a single decision tree.

### Pros and Cons of Random Forest:

#### Pros:
- Improved accuracy and robustness compared to single decision trees.
- Reduced overfitting due to the combination of multiple trees.
- Handles large datasets and high-dimensional feature spaces effectively.

#### Cons:
- Less interpretable compared to a single decision tree.
- Can be computationally expensive and require more resources to train.
- Harder to visualize the decision-making process when using multiple trees.



####In summary
 - Decision trees are a simple and interpretable approach, but they are limited by their tendency to overfit, Good for small datasets as well
 - Random Forest, as an ensemble of decision trees, addresses this limitation by combining multiple models and providing more accurate and robust predictions.



B. Techniques to Address Imbalanced Datasets:

1. Resampling Techniques:

- Oversampling: Increase the number of instances in the minority class by duplicating existing data or generating synthetic samples using techniques like SMOTE (Synthetic Minority Over-sampling Technique).

- Undersampling: Reduce the number of instances in the majority class by randomly selecting a subset of data.

2. Class Weighting:

- Assign higher weights to the minority class during model training to emphasize its importance and correct the class imbalance.

3. Ensemble Methods:
Ensemble methods like Random Forest, Gradient Boosting, or Bagging can handle imbalanced datasets better than individual models by combining predictions from multiple models, which helps improve the overall performance.

### Construct a 2-stock portfolio based on 5 years of historical adjusted close prices to construct an optimally weighted portfolio

```python
def download_historical_data(ticker):
    data = yf.download(ticker, period="5y")
    return data['Adj Close']

# Define the 2 stock tickers for the portfolio - Consumer Goods Sector
stock_tickers = ['KO', 'MCD']

# Create an empty DataFrame to store the adjusted closing prices
stock_data_df = pd.DataFrame()

# Download and store the data for each stock
for ticker in stock_tickers:
    data = download_historical_data(ticker)
    stock_data_df[ticker] = data
```

```python
# Calculate the log returns for each stock
log_returns_df = stock_data_df.pct_change().dropna()
log_returns_df = log_returns_df.apply(lambda x: pd.Series(x.dropna().values))

## Calculate num_assets
num_assets = len(log_returns_df.columns)
num_assets

#Define variable weights
weights = np.random.random(num_assets)
weights /= np.sum(weights)
weights

#weights[0] + weights[1]+ weights[2] + weights[3]+weights[4] =1

# Define the objective function to be minimized (portfolio risk)
def portfolio_risk(weights):
    return np.sqrt(np.dot(weights.T, np.dot(log_returns_df.cov() * 252, weights)))

# Define the constraint function (portfolio return)
def constraint_function(weights):
    return -np.dot(weights, log_returns_df.mean() * 252)


# Expected Portfolio Return:
def portfolio_return(weights):
    return np.sum(weights * log_returns_df.mean()) * 250

# Expected Portfolio Variance:
def portfolio_variance(weights):
    return np.dot(weights.T, np.dot(log_returns_df.cov() * 250, weights))


#Expected Portfolio Volatility:
def portfolio_Volatility (weights):
    return np.sqrt(np.dot(weights.T,np.dot(log_returns_df.cov() * 250, weights)))

# Call the functions with your weights
portfolio_return_value = portfolio_return(weights)
portfolio_variance_value = portfolio_variance(weights)
portfolio_volatility = portfolio_Volatility(weights)

# Print the results
print("Expected Portfolio Return:", portfolio_return_value)
print("Expected Portfolio Variance:", portfolio_variance_value)
print("Expected Portfolio Volatility:", portfolio_variance_value)
```

```python
## Create two empty lists. Name them pf_returns and pf_volatilites.
pf_returns = []
pf_volatilities = []
pf_weights = []
```

```python
## Create a loop with 1,000 iterations that will generate random weights, summing to 1, and will append the obtained values for the portfolio returns
## and the portfolio volatilities to pf_returns and pf_volatilities, respectively.
for x in range (1000):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)
    pf_weights.append(weights)
    pf_returns.append(np.sum(weights * log_returns_df.mean()) * 250)
    pf_volatilities.append(np.sqrt(np.dot(weights.T,np.dot(log_returns_df.cov() * 250, weights))))

pf_returns, pf_volatilities, pf_weights
```

```python
# Transform the obtained lists into NumPy arrays and reassign them to pf_returns and pf_volatilites. Once you have done that, the two objects will be NumPy arrays.
pf_returns = np.array(pf_returns)
pf_volatilities = np.array(pf_volatilities)
pf_weights = np.array(pf_weights)

pf_returns, pf_volatilities, pf_weights
```

```python
#Now, create a dictionary, called portfolios, whose keys are the strings ‚ÄúReturn‚Äù and ‚ÄúVolatility‚Äù and whose values are the NumPy arrays pf_returns and pf_volatilities

portfolios = ({'Return': pf_returns, 'Volatility': pf_volatilities})

for counter, symbol in enumerate(stock_data_df.columns.tolist()):
    #print(counter, symbol)
    portfolios[symbol+' weight'] = [w[counter] for w in pf_weights]
```

```python
portfolios = pd.DataFrame(portfolios)
portfolios.head()
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
##Calculate log_returns
logMC_returns = np.log(stocksMC_df/stocksMC_df.shift(1))
logMC_returns.head()
```
